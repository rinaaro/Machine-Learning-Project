ECONOMIC_SECTOR_VARIABLE

```python
# Define a function to map NACE codes to their nomenclature
def sector_mapping(nace_code):
    # Check if the code is in the "Industrie (et agriculture)" range (AZ to EZ)
    if "AZ" <= nace_code <= "EZ":  
        return "Industrie (et agriculture)"
    elif nace_code == "FZ":  # Exact match for FZ
        return "Construction"
    elif nace_code == "GZ":  # Exact match for GZ
        return "Commerce"
    elif nace_code == "IZ":  # Exact match for IZ
        return "HÃ©bergement et restauration"
    elif nace_code == "7820Z":  # Exact match for 7820Z
        return "IntÃ©rim"
    elif "HZ" <= nace_code <= "UZ" and nace_code != "7820Z":  # Covers HZ to UZ, excluding 7820Z
        return "Autres services hors intÃ©rim"
    else:
        return "Unknown Sector"
```

```python
learn_dataset_job["Nomenclature"] = learn_dataset_job["Economic_sector"].map(sector_mapping)
learn_nomenclature = learn_dataset_job[["PRIMARY_KEY","Nomenclature"]]
```

```python
def combine_columns(col_x, col_y):
    return col_x.fillna(col_y) if col_y is not None else col_x

for column in learn_data.columns:
    if column.endswith('_x'):
        base_column = column[:-2]  # Remove `_x` suffix
        y_column = base_column + '_y'
        if y_column in learn_data.columns:
            # Combine the columns
            learn_data[base_column] = combine_columns(learn_data[column], learn_data[y_column])
            # Drop the original `_x` and `_y` columns
            learn_data = learn_data.drop(columns=[column, y_column])
```
